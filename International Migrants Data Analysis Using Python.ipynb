```python
Sy = df[df.Geography_Name == 'Syrian Arab Republic']
```

```python
Sy
```

```python
Uk = df[df.Geography_Name == 'Ukraine']
```

```python
Uk
```

```python
plt.figure(figsize=(15,6))
plt.plot(Sy.Year, Sy.Value / 10**6)
plt.plot(Uk.Year, Uk.Value / 10**6)
plt.legend(['Syria', 'Ukraine'])
plt.xlabel('year')
plt.ylabel('value')
plt.show()
```

```python
Sy.Value
```

```python
Sy. Value / Sy.Value.iloc[0] * 100
```

```python
plt.figure(figsize=(15,6))
plt.plot(Sy.Year, Sy.Value / Sy.Value.iloc[0] * 100)
plt.plot(Uk.Year, Uk.Value / Uk.Value.iloc[0] * 100)
plt.legend(['Syria', 'Ukraine'])
plt.xlabel('year')
plt.ylabel('value growth (first year = 100)')
plt.show()
```
